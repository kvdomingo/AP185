```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.fftpack as fft
import os
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.ticker as tick
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
def read_csv(filename, save=False):
    name = filename[:-4]
    t, Vin, Vout, Vnull = np.genfromtxt(filename, delimiter=",", skip_header=2).T
    t += abs(t.min())
    rate = len(Vin)/t.max()
    c = fft.fft(Vout)
    f = np.argmax(abs(c))/len(Vin)*rate
    
    fig, (ax1, ax2) = mp.subplots(1, 2, figsize=(5*2.5, 5))
    
    ax1.plot(t, Vin, "y-")
    ax1.grid(True)
    ax1.set_ylabel(r"Input, $V_{max} = %.2f$ V"%(Vin.max()))
    ax1.xaxis.set_major_formatter(tick.NullFormatter())
    
    divider = make_axes_locatable(ax1)
    ax12 = divider.append_axes("bottom", size="100%", pad=0.1)
    
    ax12.plot(t, Vout, "c-")
    ax12.grid(True)
    ax12.set_xlabel(r"time, s")
    ax12.set_ylabel(r"Output, $V_{max} = %.2f$ V"%(Vout.max()))
    
    ax2.plot(Vin, Vout, "y.")
    ax2.grid(True)
    ax2.set_xlabel("Input, V")
    ax2.set_ylabel("Output, V")
            
    mp.suptitle("%s, %i Hz"%(name, f))
    
    if save:
        mp.savefig(name + ".png", dpi=300, bbox_inches="tight")

    mp.show()
```

```python
for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith(".csv"):
            read_csv(file, True)
```
