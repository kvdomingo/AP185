```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.signal as sig
import os
from jupyterthemes import jtplot
```

```python
%matplotlib inline
jtplot.reset()
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
```

```python
def f_cutoff(R, C):
    return 1/2/np.pi/R/C

def H(R1, R2):
    return 1 + R1/R2

def TransferSallenKeyLowPass(s, H, R1, R2, C1, C2):
    num = H/(R1*R2*C1*C2)
    den = s**2 + s*((1/R1 + 1/R2)/C1 + (1-H)/(R2*C2)) + 1/(R1*R2*C1*C2)
    return num/den

def TransferSallenKeyHighPass(s, H, R1, R2, C1, C2):
    num = H*s**2
    den = s**2 + s*((C2/R2 + C1/R2 + (1-H)*C2/R1)/(C1*C2)) + 1/(R1*R2*C1*C2)
    return num/den

def phi(s):
    return np.arctan(2*np.pi*s)
```

```python
R = 1
C = 22e-6
f_cutoff(R, C)
```

```python
f, Vout = np.genfromtxt("sallenLP.txt", skip_header=1).T
s = np.logspace(0, 7, 500, base=10)
w = 2*np.pi*s

mp.plot(2*np.pi*f, 20*np.log10(Vout/5), "ko")
mp.plot(w, 20*np.log10(abs(TransferSallenKeyLowPass(1j*w, H(R, R), R, R, C, C))), "b--")
mp.xscale("log")
mp.xlabel("$\omega$")
mp.ylabel("dB")
mp.grid(which="both", axis="x", linestyle="--")
mp.grid(axis="y", which="major", linestyle="--")
mp.show()
```

```python
f, Vout = np.genfromtxt("sallenHP.txt", skip_header=1).T

mp.plot(2*np.pi*f, 20*np.log10(Vout/5), "ko")
mp.plot(s, 20*np.log10(abs(TransferSallenKeyHighPass(1j*w, H(R, R), R, R, C, C))), "b--")
mp.xscale("log")
mp.xlabel("$\omega$")
mp.ylabel("dB")
mp.grid(which="both", axis="x", linestyle="--")
mp.grid(axis="y", which="major", linestyle="--")
mp.show()
```

```python
def H(R3, R4):
    return 1 + R4/R3

def TransferMultipleLow(s, H, R1, C2, R3, R4, C5):
    num = -H/(R1*R3*C2*C5)
    den = s**2 + s/C2*(1/R1 + 1/R3 + 1/R4) + 1/(R3*R4*C2*C5)
    return num/den

def TransferMultipleHigh(s, H, C1, R2, C3, C4, R5):
    num = -s**2 * C1/C4
    den = s**2 + s*((C1+C3+C4)/C3/C4/R5) + 1/R2/R5/C3/C4
    return num/den
```

```python
f, Vout = np.genfromtxt("MFlow.txt", skip_header=1).T

mp.plot(2*np.pi*f, 20*np.log10(Vout/5), "ko")
mp.plot(s, 20*np.log10(abs(TransferMultipleLow(1j*w, H(R, R), R, C, R, R, C))), "b--")
mp.xscale("log")
mp.xlabel("$\omega$")
mp.ylabel("dB")
mp.grid(which="both", axis="x", linestyle="--")
mp.grid(axis="y", which="major", linestyle="--")
mp.show()
```

```python
f, Vout = np.genfromtxt("MFhigh.txt", skip_header=1).T

mp.plot(2*np.pi*f, 20*np.log10(Vout/5), "ko")
mp.plot(s, 20*np.log10(abs(TransferMultipleHigh(1j*w, H(R, R), C, R, C, C, R))), "b--")
mp.xscale("log")
mp.xlabel("$\omega$")
mp.ylabel("dB")
mp.grid(which="both", axis="x", linestyle="--")
mp.grid(axis="y", which="major", linestyle="--")
mp.show()
```
